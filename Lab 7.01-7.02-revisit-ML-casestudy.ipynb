- Check for null values in all the columns

```python
# choose categorical columns
categorical = data.select_dtypes(object)
categorical.head()
```

```python
# filling empty cells with NaN
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
#categorical.isna().sum()/len(categorical)
```

```python
nulls_percent_df_cat = pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
nulls_percent_df_cat.columns = ['column_name', 'nulls_percentage']
nulls_percent_df_cat
```

- Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:

    OSOURCE - symbol definitions not provided, too many categories
    ZIP CODE - we are including state already
- Identify columns that over 85% missing values

- Remove those columns from the dataframe

```python
drop_list = ['ZIP','OSOURCE']
```

```python
columns_above_threshold_cat = nulls_percent_df_cat[nulls_percent_df_cat['nulls_percentage']>0.85]
```

```python
drop_columns_list = list(columns_above_threshold_cat['column_name'])
print(drop_columns_list)
len(drop_columns_list)
```

```python
cols = ['VETERANS','SOLIH']
for item in cols:
    drop_columns_list.remove(item)
print(drop_columns_list) 
```

```python
drop_columns_list = drop_columns_list + drop_list
```

```python
categorical = categorical.drop(columns=drop_columns_list, axis=1)
```

- Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
categorical['GENDER'].value_counts(dropna=False)
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
categorical['GENDER'] = np.where(categorical['GENDER'].isin(['J','C','A','U']) , 'other', categorical['GENDER'])
```
